```python
def books_alike(mybook, k):
    mybook = [mybook]

    summ = dists[mybook].sum(axis=1)
    summ = summ.order(ascending=False)
  
    ranked_books = summ[summ.index.isin(mybook)==False]

    ranked_books = pd.DataFrame(ranked_books, index=None, columns=[['Similarity']])

    book_names = ranked_books.join(booknames, how='inner')

    book_names_sorted = book_names.sort(['Similarity'], ascending=False)
        
    mybookname = booknames[booknames.index==mybook]['Book-Title']
    
    print 'Book I like:', mybookname
    return book_names_sorted.head(k)
    
```

#Function to identify similar users

```python
n=1000

top_n = allbooks['User-ID'].value_counts().index[:n]

df = allbooks[allbooks['User-ID'].isin(top_n)]

# allbooks[allbooks.ISBN=='0671024248'].ISBN.value_counts()

last_n = allbooks['User-ID'].value_counts().index[n-1:n]

print 'Minimum number of ratings', allbooks[allbooks.ISBN.isin(last_n)].ISBN.value_counts()
print 'Minimum number of ratings', allbooks[allbooks['User-ID'].isin(last_n)]['User-ID'].value_counts()
```

```python
df_pivot = pd.pivot_table(df, values=['Book-Rating'], columns=['User-ID', 'ISBN'], aggfunc = np.mean )

df_wide = df_pivot.unstack(-1)

df_wide.head(10)

df_wide.shape

df_wide = df_wide.fillna(0)

UserID = df_wide.index.levels[1]

df_wide.index.levels[0]

df_wide.columns

UserID
```

```python
from sklearn.metrics.pairwise import cosine_similarity

dists = cosine_similarity(df_wide)
```

```python
dists = pd.DataFrame(dists, index=UserID, columns=UserID)

```

```python
#how do you get it to show 1000 rows by 1000 columns

dists.ix[:10000, :10000]

dists.head(100)
```

```python
def users_alike(myuser, k):
    myuser = [myuser]

    summ = dists[myuser].sum(axis=1)
    summ = summ.order(ascending=False)
  
    ranked_users = summ[summ.index.isin(myuser)==False]

    ranked_users = pd.DataFrame(ranked_users, index=None, columns=[['Similarity']])

    user_info = ranked_users.join(allusers, how='inner')

    user_info_sorted = user_info.sort(['Similarity'], ascending=False)
        
    myusername = allusers[allusers.index==myuser]
    
    print 'Book I like:', myusername
    return user_info_sorted.head(k)
    
```

```python
users_alike(30972, 10)
```
